```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='darkgrid')
```

```python
#데이터셋
tips=sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
#hue 색상
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker')
```

```python
#hue 색상
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker')
```

```python
#style & hue
sns.relplot(x='total_bill',y='tip',data=tips,style='smoker',hue='size')
```

```python
# 색상
sns.relplot(x='total_bill',y='tip',data=tips,hue='size',palette='Blues_d')
```

```python
#size
sns.relplot(x='total_bill',y='tip',data=tips,size='size',palette='Blues_d')
```

```python
df=pd.DataFrame(dict(time=np.arange(500),value=np.random.rand(500).cumsum()))
df
```

```python
sns.relplot(x='time',y='value',data=df,kind='line')
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint',y = 'signal',kind='line',data=fmri)
```

```python
sns.relplot(x='timepoint',y = 'signal',kind='line',data=fmri,ci=None)
```

```python
sns.relplot(x='timepoint',y = 'signal',kind='line',data=fmri,ci='sd')
```

```python
sns.relplot(x='timepoint',y = 'signal',kind='line',data=fmri, hue = 'region',style='event')
```
