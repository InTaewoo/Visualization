```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 지정후 깨지는거 방지
%matplotlib inline
```

```python
total_car =[2977000,2969000,2974000,3014000,3057000,3083000,3116000,3125000,3124000,3157000]
total_parking=[3595000,3584000,3760000,3822000,3877000,3983000,4053000,4130000,4251000,4328000]
year = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
index = np.arange(len(year))
```

```python
# 꺾은선 그래프
plt.figure(figsize=(11,5))
plt.plot(index,total_car,color='green',marker='o',linestyle='solid')
plt.plot(index,total_parking,color='blue',marker='o',linestyle='solid')
plt.legend(['자동차등록대수','주차장 수'],loc=0)
plt.xlabel('연도')
plt.ylabel('자동차/주자창수')
plt.xticks(index,year)
plt.title('연도별 서울시 자동차 등록대수 및 주차장 수')
plt.grid(True)
plt.show()
```

```python
# 막대그래프
plt.figure(figsize=(10,5))
plt.bar(index,total_car,align='edge',color='green',width=0.3)
plt.bar(index+0.3,total_parking,align='edge',color='blue',width=0.3)
plt.legend(['자동차등록대수','주차장 수'],loc=0)
plt.xlabel('연도')
plt.ylabel('자동차/주자창수')
plt.title('서울시 연도별 자동차 등록대수 및 주차장 수')
plt.xticks(index+0.3,year)
plt.grid(True)
plt.show()
```
