```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 지정후 깨지는거 방지
%matplotlib inline
```

```python
total = pd.read_csv('서울시 전체 주차장확보율.csv')
total
```

```python
#2020년의 자치구
gu = ['중구','종로구','강남구','마포구','금천구','용산구','동대문구','서초구','강서구','광진구','강동구','서대문구','송파구','성북구','구로구','동작구','관악구','영등포구','성동구','은평구','강북구','노원구','중랑구',
      '도봉구','양천구']
```

```python
percent = gu['주차장확보율'][1:26].unique()
percent_sort=sorted(percent,reverse=True)
percent_sort
```

```python
index = np.arange(len(gu))
index
```

```python
# 막대그래프
plt.figure(figsize=(17,5))
plt.bar(index,percent_sort,color='green',width=0.3)

plt.legend(['주차장 확보율'],loc=0)
plt.xlabel('자치구')
plt.ylabel('%')
plt.title('지역구 별 주차장 확보율',fontsize=25)
plt.xticks(index,gu)
plt.grid(True)
plt.show()
```
